```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, KFold
import warnings
warnings.filterwarnings("ignore")
veri_seti = pd.read_excel('chronic_kidney_disease.xlsx')
# Veri setini yükleyin (veri setinin adını ve yolunu güncelleyin)



# Bağımsız değişkenler ve hedef değişkeni ayırın
X = veri_seti.drop(columns=['class'])
y = veri_seti['class']

# K-Fold çapraz doğrulama için KFold nesnesini oluşturun
kf = KFold(n_splits=5, shuffle=True, random_state=0)

# Sınıflandırma modelini oluşturun (Logistic Regression kullanıldı, diğer modeller de kullanılabilir)
#Bu satır, LogisticRegression sınıfının kurucu işlevine geçirilen parametrelerle bir lojistik regresyon modeli oluşturur.
#Bu modeli daha sonra veriye uygulamak ve sınıflandırma görevleri için kullanmak mümkün olacaktır.
model = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=1000)

# K-Fold çapraz doğrulama ile model performansını değerlendirin
scores = cross_val_score(model, X, y, cv=kf)

# Her katmanın performans sonuçlarını yazdırın
for i, score in enumerate(scores):
    print(f"Katman {i + 1} Skoru: {score:.2f}")

# Ortalama performans sonucunu yazdırın
print(f"Ortalama Skor: {scores.mean():.2f}")
```
